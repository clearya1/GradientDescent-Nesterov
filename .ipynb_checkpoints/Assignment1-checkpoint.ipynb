### Gradient descent

$$
x_{n+1}=x_{n}-h \nabla f(x_{n})
$$

#### Convex
$ h= \frac{1}{L} $

#### Strongly Convex
$ h=\frac{2}{\mu+L}$


### Nesterov 

\begin{align}
x_{k+1} &= y_{k}-\frac{1}{L} \nabla f(y_{k}) \\
y_{k+1} &= x_{k+1}+\gamma_{k}(x_{k+1}-x_{k})
\end{align}

#### Convex
$$\gamma_{k} = \frac{t_{k}-1}{t_{k+1}}, \quad t_{k+1}=\frac{1}{2} \left(1+\sqrt{1+4t^{2}_{k}} \right), \quad t_{-1}=1 $$

#### Strongly Convex
$$ \gamma_{k} = \frac{\sqrt{L}-\sqrt{\mu}}{\sqrt{L}+\sqrt{\mu}} $$ 




# Question 1: Polynomial Regression

Consider the least squares regression problem introduced in lectures, where we wish to reconstruct an unknown function from noisy point values. Suppose we are given a set of points $\{u_i, y_i\}_{i=1}^m$, and we want to find the degree $n-1$ polynomial that best fits these points.  In other words, we want to learn the unknown coefficients $\{x_j\}_{j=1}^n$ in the polynomial
$$
f(u; x) = \sum_{j=1}^{n} {x_j} u^{j-1}, 
$$
such that the residual $\sum_{i=1}^m (y_i - f(u_i;x))^2$ is minimised.


### 1.1

Show that the regression problem above can be formulated as the linear inverse problem

$$
x^* = \text{argmin} \frac{1}{2} \|A x - y\|_2^2,
$$

where $y = \{y_i\}_{i=1}^m \in \mathbb R^m$ and $A \in \mathbb R^{m \times n}$ is the Vandermonde matrix with entries $a_{ij} = u_i^{j-1}$.

*Enter your solution here*

We are aiming to minimise the residual with respect to the coefficients $\{x_j\}_{j=1}^n$. This means we are trying to find the optimum values for $\{x_j\}_{j=1}^n = x^*$ such that the right hand side of the following equation is minimised:
$$x^* = \text{argmin}_{x_j} \sum_{i=1}^m\left( y_i - \sum_{j=1}^n x_ju_i^{j-1}  \right)^2$$

We can sub in $A \in \mathbb R^{m \times n}$, the Vandermonde matrix with entries $a_{ij} = u_i^{j-1}$, to get:
$$x^* = \text{argmin}_{x} \sum_{i=1}^m\left( y_i - \sum_{j=1}^n a_{ij}x_j  \right)^2$$

$$x^* = \text{argmin}_{x} \sum_{i=1}^m\left( y_i - (Ax)_i  \right)^2$$

We now see that this is just the definition of the square norm of the vector $y - Ax$, which equals the squared norm of the vector $Ax - y$:

$$x^* = \text{argmin}_{x} \|Ax-y\|^2_2$$

Finally, to match the expression given in the question, we note that multiplying the expression on the right hand side by a constant scalar, 1/2, will not make a difference when we take the argmin. This is because whichever $x$ is the minimiser of the expression, will also be the minimiser of this constant scaling of the expression:
$$x^* = \text{argmin}_{x} \frac{1}{2}\|Ax-y\|^2_2$$

### 1.2

Show that the gradient of $\frac{1}{2} \|A x - y\|_2^2$ with respect to $x$ is 

$$A^\mathrm T A x - A^\mathrm T y.$$

*Enter your solution here*

First, we get rid of the norm so that the expression is easier to work with:

\begin{align}
\frac{1}{2}||Ax-y||^2_2 &= \frac{1}{2}\sum_{i=1}^m\left( y_i - (Ax)_i \right)^2 \\
&= \frac{1}{2}\sum_{i=1}^m\left( y_i - \sum_{j=1}^na_{ij}x_j \right)^2
\end{align}

Now, take the gradient with respect to a single coefficient, $x_k$. Note that $\frac{\partial x_j}{\partial x_k} = \delta_{jk}$:

\begin{align}
\frac{\partial}{\partial x_k}\left(  \frac{1}{2}\sum_{i=1}^m\left(\sum_{j=1}^na_{ij}x_j - y_i \right)^2 \right) &= \sum_{i=1}^m\left(\sum_{j=1}^na_{ij}x_j - y_i \right)a_{ij}\delta_{jk} \\
&= \sum_{i=1}^m\left(\sum_{j=1}^na_{ij}x_j - y_i \right)a_{ik} \\
&= \sum_{i=1}^m\left((Ax)_i - y_i\right)a_{ik} \\
&= \sum_{i=1}^ma_{ik}\left((Ax)_i - y_i\right) \\
&= (A^\mathrm T A x)_k - (A^\mathrm T y)_k
\end{align}

In the second last line, we used that each $a_{ij}$ is just a scalar so it is commutative under multiplication. In the final line, we used the following equations. Note the difference in the positioning of the index that we are summing over as this distinguishes $A^\mathrm T$ from $A$: 
$$(A^\mathrm T x)_k = \sum_{i=1}^m a_{ik}x_i$$       $$(A x)_k = \sum_{i=1}^n a_{ki}x_i$$

Thus, as $\nabla_k = \frac{\partial}{\partial x_k}$:
$$\nabla_k\left(\frac{1}{2} \|A x - y\|_2^2\right) = (A^\mathrm T A x)_k - (A^\mathrm T y)_k$$
Finally, getting rid of the index $k$:
$$\nabla\left(\frac{1}{2} \|A x - y\|_2^2\right) = A^\mathrm T A x - A^\mathrm T y$$
as required.

### Overview of the Code

The main part of the functions implementing the Nesterov and gradient descent methods (`gradDesc`,`nesterov` and `nesterovStrong`) is a `while` loop, which runs until all of the elements of the gradient are sufficiently close to 0. This tolerance is set by both `atol` and `rtol` by the following equation, so that the code stops running when the following equality is true: 

$$|\text{gradient}| \le \text{atol} + \text{rtol} * |\text{gradient}|$$ 

This equation can be found in the documentation of the [numpy.allclose](https://numpy.org/doc/stable/reference/generated/numpy.allclose.html#numpy-allclose) function.

If the boolean `metrics` is set to `True`, then the functions implementing the Nesterov and gradient descent methods (`gradDesc`,`nesterov` and `nesterovStrong`) record and return the Euclidean norm of the gradient at each iteration. This is done so that we can track how quickly each solution converges to a solution. We used the `numpy.linalg.norm` function to compute this, and its documentation can be found [here](https://numpy.org/doc/stable/reference/generated/numpy.linalg.norm.html). The smaller the norm of the gradient vector, the closer we are to the minimum of the residual function, $f$. We plot the norm as it is a scalar quantity and so is easy to plot against the iteration number. We could also plot the residuals function, $f$, as a function of the iteration number to see how the methods converge, and it would be easy to modify the code to do this. However, I chose to plot the norm of the gradient vector because we know that the true value of this should be equal to 0, and because it is related to my stopping criterion, mentioned at the start of this cell. 

Another point is that the functions implementing the Nesterov and gradient descent methods (`gradDesc`,`nesterov` and `nesterovStrong`) do not take the matrix, $A$ as an input. Instead, they take an array of matrices equal to ($A^\mathrm T A$, $A^\mathrm T y$). This is done to speed up the code, so that we do not have to compute these constant values at every iteration of the methods.

Note also that the three functions, `gradDesc`,`nesterov` and `nesterovStrong`, take an input `grad`. This refers to which function we should use for computing the gradient as in Ex 2, we will introduce a regularisation term which will change the gradient function slightly. Otherwise, the code will be unchanged though so this is a nice way of not having to rewrite the entire functions. 

Otherwise, the functions `gradDesc`,`nesterov` and `nesterovStrong` simply implement their respective methods, according to the equations in our notes, and the start of the Jupyter notebook.

### Computing $L$ and $\mu$

In order to compute $L$ and $\mu$, we can quickly compute the Hessian of our residual function, $f$ to be the following: 

$$\nabla\nabla f = A^\mathrm T A$$ 

This follows trivially from Ex 1.3, where we computed the gradient of the residual function, $\nabla f = A^\mathrm T A x - A^\mathrm T y$. Once we have the Hessian matrix, we use the fact that 

$$\lambda_n I \le \nabla\nabla f \le \lambda_1 I$$ 

where $\lambda_1 \ge \dots \ge \lambda_n$ are the eigenvalues of the Hessian matrix, $A^\mathrm T A$. From this, it follows that $L = \lambda_1$, the largest eigenvalue, and $\mu = \lambda_n$, the smallest eigenvalue. Note that for strongly convex problems, we have that $\lambda_n > 0$. 

In the code, I used the `numpy.linalg` package to compute the eigenvalues of the Hessian matrix and set the largest one equal to $L$ and the smallest one equal to $\mu$. 

```python
import numpy as np

# add code here

"""define necessary numerical constants"""

rtol = 1e-5      #relative tolerance - determines how close the gradient should be to 0 at the minimiser
atol = 1e-9      #absolute tolerance - determines how close the gradient should be to 0 at the minimiser
                 #                     see formula for tolerance above

"""define initial estimate for the coefficients"""

x0 = np.array([0.,0.])    #set initial estimate equal to array of zeros


def grad1(As, x, y):
    
    """compute the derivative of the residual, according to the formula from 1.2"""
    """inputs are an array with [A^T.A, A^T.y], current estimate for co-efficients (x), target data (y)"""
    """returns the gradient"""
    
    return np.matmul(As[0],x) - As[1]      #return A^T.A.x - A^T.y


def gradDesc(As, x0, u, y, h, grad, metrics=False):
    
    """implement the full gradient descent method"""
    """inputs are an array with [A^T.A, A^T.y], initial estimate for co-efficients (x0), input data (u and y), the step size (h), the gradient function (grad) and a boolean"""
    """returns the numerical minimiser (x^*) and how the gradients improve with iterations if metrics = True"""
    
    x = x0                                 #load the initial estimate for coefficients
    gradient = grad(As, x, y)              #load the initial value for the gradient - note this could also be initialised to array of 1s here
    targetGrad = np.zeros_like(gradient)   #create a matrix of 0s like the gradient as...
                                           #...want every component of the gradient to be sufficiently close to zero
    
    if metrics==True:                      #if storing the performance of the algorithm...
        gradientDists = []                 #...create an array to store the Euclidean norms of the gradient
    
    while not (np.allclose(targetGrad, gradient, rtol, atol)):   #Stops if absolute(gradient) <= (atol + rtol * absolute(gradient))
                                                                 #will keep iterating until the gradient of the residual... 
                                                                 #...is sufficiently close to zero
        
        gradient = grad(As, x, y)          #compute next value of gradient
        x -= h*gradient                    #compute the next iteration of x
        
        if metrics==True:                                    #if storing the performance of the algorithm...
            gradientDists.append(np.linalg.norm(gradient))   #...store the Euclidean norms of the gradient
    
    print("Final gradient : ", gradient)   #print the final value for the gradient - should be very close to 0 
    
    if metrics==False:                     #if we don't want the performance...     
        return x                           #...return the minimiser
    else:                                  #if we do want the performance as well...
        return x, gradientDists            #...return the Euclidean norm of the gradients at each iteration as well
        
# Testing the function on a small example
u = np.array([-0.5, 0, 0.5])
y = np.array([0., 1., 2.])
A = np.array([[1., -0.5],[1., 0.],[1., 0.5]])

AtA = np.matmul(A.T, A)             #compute the constant value of A^T.A once here
Aty = np.matmul(A.T, y)             #compute the constant value of A^T.y once here

"""determine a suitable h for this strongly convex problem"""

hess = AtA                          #the Hessian matrix just equals to A^T.A
eigs = np.linalg.eigvals( hess )    #compute the eigenvalues of the Hessian matrix
L = np.max( eigs )                  #Lipschitz gradient constant is the maximum eigenvalue
mu = np.min( eigs )                 #continuously differentiable constant for strong convexity is the minimum eigenvalue

h = 2./(L+mu)                       #strongly convex formula for h as the problem is overdetermined (m=n)

print("Minimiser equals : ", gradDesc([AtA, Aty], x0, u, y, h, grad1))   #compute and print the minimiser using the gradient descent method
                                                                         #using the standard formula for the gradient (grad1)
```

### 1.4

In the code cell below, implement the Nestorov method both for convex and strongly convex objectives and  compute a minimiser $x_{*}$. 

Test your method on the simple, noise-free example given at the bottom of the code cell, to which the answer is $x = \begin{bmatrix} 1 \\ 2 \end{bmatrix}$.

```python
import numpy as np

# add code here

"""define necessary numerical constants"""

rtol = 1e-5      #relative tolerance - determines how close the gradient should be to 0 at the minimiser
atol = 1e-9      #absolute tolerance - determines how close the gradient should be to 0 at the minimiser
                 #                     see formula for tolerance above

def nesterov(As, x0, u, y, L, grad, metrics=False):
    
    """implement the full Nesterov method for convex functions"""
    """inputs are an array with [A^T.A, A^T.y], initial estimate for co-efficients (x0), input data (u and y), the Lipschitz constant (L), the gradient function (grad) and a boolean"""
    """returns the numerical minimiser (x^*) and how the gradient improves with iterations if metrics=True"""
    
    ta = 0.5*(1+np.sqrt(1+4))              #now need to keep track of two t values
    tb = 0.5*(1+np.sqrt(1+4*ta**2))        #load ta = t0, tb = t1 to initialise sequence of t values
    
    gamma = (ta - 1)/tb                    #load initial value of gamma = (t0 - 1)/t1
    
    x = x0                                 #load the initial estimates for x0 and z0 (z_k = y_k from notes as y already denotes the input data)
    z = x0                                 #can see from formulae that gamma_{-1} = 0 so x0 = z0
    
    gradient = grad(As, z, y)              #load the initial value for the gradient - note this could also be initialised to array of 1s here
    targetGrad = np.zeros_like(gradient)   #create a matrix of 0s like the gradient as...
                                           #...want every component of the gradient to be sufficiently close to zero
    
    if metrics==True:                      #if storing the performance of the algorithm...
        gradientDists = []                 #...create an array to store the Euclidean norms of the gradient
    
    while not (np.allclose(targetGrad, gradient, rtol, atol)):   #Stops if absolute(gradient) <= (atol + rtol * absolute(gradient))
                                                                 #will keep iterating until the gradient of the residual... 
                                                                 #...is sufficiently close to zero
        
        gradient = grad(As, z, y)          #compute next value of gradient
                                           #now update the sequences x and z
        xNew = z - 1./L*gradient           #cannot just overwrite x with next iteration as the sequence z...
        z = xNew + gamma*(xNew - x)        #...depends on both x_{k} and x_{k+1}
        
        x = xNew                           #now overwrite x with xNew
        
        ta = tb                            #shift the sequence of t values along so t_{k} <- t_{k+1}
        tb = 0.5*(1+np.sqrt(1+4*ta**2))    #update t_{k+1}
        
        gamma = (ta - 1)/tb                #update the value of gamma
        
        if metrics==True:                                    #if storing the performance of the algorithm...
            gradientDists.append(np.linalg.norm(gradient))   #...store the Euclidean norm of the gradient
    
    print("Convex final gradient : ", gradient)   #print the final value for the gradient - should be very close to 0 
    
    if metrics==False:                     #if we don't want the performance...     
        return x                           #...return the minimiser
    else:                                  #if we do want the performance...
        return x, gradientDists            #...return the Euclidean norms of the gradients at each iteration as well

def nesterovStrong(As, x0, u, y, L, mu, grad, metrics=False):
    
    """implement the full Nesterov method for strongly convex functions"""
    """inputs are an array with [A, A^T.A, A^T.y], initial estimate for co-efficients (x0), input data (u and y), Lipschitz constant (L), cont. diff. constant (mu), the gradient function (grad) and a boolean"""
    """returns the numerical minimiser (x^*) and how the gradients improve with iterations if metrics=True"""
       
    gamma = (np.sqrt(L)-np.sqrt(mu))/(np.sqrt(L)+np.sqrt(mu))   #calculate the constant value of gamma
    
    x = x0                                 #load the initial estimates for x0 and z0 (z_k = y_k from notes as y already denotes the input data)
    z = x0                                 #can see from formula that gamma_{-1} = 0 so x0 = z0
    
    gradient = grad(As, z, y)              #load the initial value for the gradient - note this could also be initialised to array of 1s here
    targetGrad = np.zeros_like(gradient)   #create a matrix of 0s like the gradient as...
                                           #...want every component of the gradient to be sufficiently close to zero
    
    if metrics==True:                      #if storing the performance of the algorithm...
        gradientDists = []                 #...create an array to store the Euclidean norms of the gradient
    
    while not (np.allclose(targetGrad, gradient, rtol, atol)):   #Stops if absolute(gradient) <= (atol + rtol * absolute(gradient))
                                                                 #will keep iterating until the gradient of the residual... 
                                                                 #...is sufficiently close to zero
        
        gradient = grad(As, z, y)          #compute next value of gradient
                                           #now update the sequences x and z
        xNew = z - 1./L*gradient           #cannot just overwrite x with next iteration as the sequence z...
        z = xNew + gamma*(xNew - x)        #...depends on both x_{k} and x_{k+1}
        
        x = xNew                           #now overwrite x with xNew
        
        if metrics==True:                                    #if storing the performance of the algorithm...
            gradientDists.append(np.linalg.norm(gradient))   #...store the Euclidean norms from 0 of the gradient
    
    print("Strongly Convex final gradient : ", gradient)     #print the final value for the gradient - should be very close to 0 
    
    if metrics==False:                     #if we don't want the performance...     
        return x                           #...return the minimiser
    else:                                  #if we do want the performance...
        return x, gradientDists            #...return the Euclidean norms of the gradients at each iteration as well

# Testing the function on a small example
u = np.array([-0.5, 0, 0.5])
y = np.array([0., 1., 2.])
A = np.array([[1., -0.5],[1., 0.],[1., 0.5]])

AtA = np.matmul(A.T, A)            #compute the constant value of A^T.A once here
Aty = np.matmul(A.T, y)            #compute the constant value of A^T.y once here

"""determine suitable L and mu for this strongly convex problem"""

hess = AtA                         #the Hessian matrix just equals to A^T.A
eigs = np.linalg.eigvals( hess )   #compute the eigenvalues of the Hessian matrix
L = np.max( eigs )                 #Lipschitz gradient constant is the maximum eigenvalue
mu = np.min( eigs )                #continuously differentiable constant for strong convexity is the minimum eigenvalue

print("Nesterov minimser for convex : ", nesterov([AtA, Aty], np.zeros(2), u, y, L, grad1))                    #compute and print the minimiser using Nesterov's method for convex functions
print("Nesterov minimser for strongly convex : ", nesterovStrong([AtA, Aty], np.zeros(2), u, y, L, mu, grad1)) #compute and print the minimiser using Nesterov's method for strongly convex functions
                                                                                                               ##using the standard formula for the gradient (grad1)
```

# Question 2: Regularised Polynomial Regression 

When $m \leq n$ (i.e. we are dealing with an underdetermined least squares problem), the minimisation problem in Question 1.1 is convex.

Consider now the regularised minimisation problem

$$
x_* = \text{argmin} \frac{1}{2} \|A x - y\|_2^2 + \frac{\lambda}{2} \| x\|_2^2,
$$

which is strongly convex, and hence has a unique solution $x_*$, for any choice of regularisation parameter $\lambda > 0$.

### 2.1

Show that the gradient of $\frac{1}{2} \|A x - y\|_2^2 + \frac{\lambda}{2} \| x\|_2^2$ with respect to $x$ is 

$$(A^\mathrm T A + \lambda I) x - A^\mathrm T y.$$

*Add your solution here*

Due to the linearity of the gradient operator, we can write:

$$
    \nabla{\frac{1}{2} \|A x - y\|_2^2 + \frac{\lambda}{2} \| x\|_2^2} = \nabla{\frac{1}{2} \|A x - y\|_2^2} + \nabla{\frac{\lambda}{2} \| x\|_2^2}
$$

Now we can just sub in our expression from question 1.2, leaving us to evaluate the second term in the equation above. Using a similar logic from question 1.2, where we compute the gradient element-wise:

\begin{align}
\nabla_{k}\left(\frac{\lambda}{2} \| x\|_2^2\right) &= \frac{\partial}{\partial x_k}\frac{\lambda}{2} \| x\|_2^2 \\
&= \frac{\partial}{\partial x_k}\frac{\lambda}{2}\sum_{i=1}^n(x_i)^2 \\
&= \lambda\sum_{i=1}^n(x_i)\delta_{ik} \\
&= \lambda x_{k}
\end{align}

Thus, writing the above equation in vector notation: $\nabla{\frac{\lambda}{2} \| x\|_2^2} = \lambda x$. We can now sub this result and our expression from question 1.2 into our full expression:

\begin{align}
\nabla{\frac{1}{2} \|A x - y\|_2^2 + \frac{\lambda}{2} \| x\|_2^2} &= A^\mathrm T A x - A^\mathrm T y + \lambda x \\
&= (A^\mathrm T A + \lambda I) x - A^\mathrm T y
\end{align}

as required.

### Numerically showing that this is a Lyapunov function

We show that $V(x_{k},x_{k-1})$ is a Lyanpunov function in the following code cell. We use the residual function with the regularisation term as it is a strongly convex function. In order to evaluate $V(x_{k},x_{k-1})$, we need a numerical estimate for $x^*$ and thus $f(x^*)$. Thus, we use the result of the Ex 2.3 with the Nesterov method for strongly convex function as we found that it was the most efficient method. 

We implement the Nesterov method for strongly convex functions again, setting the same tolerances as in the previous section so that we generate the same solution. But now, at every iteration, we evaluate the function $V(x_{k},x_{k-1})$ using the $x^*$ calculated previously. We append it to a list so that we can analyse if $V(x_{k+1},x_{k}) \leq V(x_{k},x_{k-1})$ holds true.

Then to show that this holds true, we can iterate through this list and check that each iteration of $V$ is smaller than the previous iteration. A nice way to show this visually is to plot $\frac{V(x_{k+1},x_{k})}{V(x_{k},x_{k-1})}$ and show that this quantity is always smaller than 1. Both of these checks are implemented in the following code cell and the results are shown below. 

### Deducing a convergence rate 

Let's start with this inequality and write it out fully. We will eventually be able to telescope this inequality and find a convergence rate which depends on the iteration number, $k$.

$$V(x_{k},x_{k-1}) \leq V(x_{k-1},x_{k-2})$$

$$(1-\delta)^{-k}\left(f(x_{k})-f(x_{*}) +\frac{\mu}{2}\left\|\frac{(1-\delta)}{\delta}(x_{k}-x_{k-1})+x_{k}-x_{*} \right\|^{2} \right) \le (1-\delta)^{-(k-1)}\left(f(x_{k-1})-f(x_{*}) +\frac{\mu}{2}\left\|\frac{(1-\delta)}{\delta}(x_{k-1}-x_{k-2})+x_{k-1}-x_{*} \right\|^{2} \right)$$

Now we can divide both sides by $(1-\delta)^{-k}$:

$$f(x_{k})-f(x_{*}) +\frac{\mu}{2}\left\|\frac{(1-\delta)}{\delta}(x_{k}-x_{k-1})+x_{k}-x_{*} \right\|^{2} \le (1-\delta)\left(f(x_{k-1})-f(x_{*}) +\frac{\mu}{2}\left\|\frac{(1-\delta)}{\delta}(x_{k-1}-x_{k-2})+x_{k-1}-x_{*} \right\|^{2} \right)$$

To make this easier to follow, I will drop the contents of the norm. I will write the norm out fully again at the end of the proof:

$$f(x_{k})-f(x_{*}) +\frac{\mu}{2}\left\|\cdot \right\|^{2} \le (1-\delta)\left(f(x_{k-1})-f(x_{*}) +\frac{\mu}{2}\left\|\cdot \right\|^{2} \right)$$

We can now just telescope this inequality as follows:

\begin{align}
f(x_{k})-f(x_{*}) +\frac{\mu}{2}\left\|\cdot \right\|^{2} &\le (1-\delta)^2\left(f(x_{k-2})-f(x_{*}) +\frac{\mu}{2}\left\|\cdot \right\|^{2} \right) \\
&\vdots \\
&\le (1-\delta)^{k-1}\left(f(x_{1})-f(x_{*}) +\frac{\mu}{2}\left\|\cdot \right\|^{2} \right)
\end{align}

Thus, writing the norm out fully, and isolating $f(x_{k})-f(x_{*})$, we get the following rate of convergence for $f(x_{k})-f(x_{*})$:

$$f(x_{k})-f(x_{*}) \le (1-\delta)^{k-1}\left(f(x_{1})-f(x_{*}) +\frac{\mu}{2}\left\| \frac{(1-\delta)}{\delta}(x_{1}-x_{0})+x_{1}-x_{*} \right\|^{2} \right) - \frac{\mu}{2}\left\|\frac{(1-\delta)}{\delta}(x_{k}-x_{k-1})+x_{k}-x_{*} \right\|^{2} $$

Note that we cannot go any further in our telescoping as we would then have an unknown quantity, $x_{-1}$, in our inequality. Finally, we can simplify this slightly by getting rid of the final term if we want. This is true because it is non-negative and we are subtracting it from the greater side of the inequality, so the inequality still holds if we get rid of it:

$$f(x_{k})-f(x_{*}) \le (1-\delta)^{k-1}\left(f(x_{1})-f(x_{*}) +\frac{\mu}{2}\left\| \frac{x_1}{\delta} + \frac{(\delta-1)x_0}{\delta}-x_{*} \right\|^{2} \right)$$

This means that the rate of convergence is linear as the rate is given in terms of an exponential function of the iteration counter, $k$. In other words, at each iteration, the difference $f(x_{i})-f(x_{*})$ is multiplied by a constant which equals $(1-\delta)$. Thus, it is a linear rate of convergence. This also means that the rate is fast as each new right digit of the answer takes a constant amount of computations. 

```python
import numpy as np

# add code here

def f(A, x, y):
    
    """calculate the function f in the V function"""
    """inputs are the matrix (A) and the coefficients (x)"""
    """returns \frac{1}{2} \|A x - y\|_2^2 + \frac{\lambda}{2} \| x\|_2^2 """
    
    Ax = np.matmul(A,x)          #compute the matrix product of A.x
    
    return 0.5*np.sum( (Ax-y)**2 ) + lam/2. * np.sum( x**2 ) #return f


def V(A, xk2, xk1, k, xmin, fmin, y):
    
    """calculate V(x_k, x_{k-1})"""
    """inputs are the matrix (A), the two iterations of coefficients (xk2, xk1), the index (k), the numerical minimiser (xmin), the minimum of the function f (fmin) and the data (y)"""
    """returns V(xk2, xk1)"""
    
    normsq = np.sum( ( (1-delta)/delta * (xk2 - xk1) + xk2 - xmin )**2 )  #compute the normed term in the V function
    
    return (1-delta)**(-k) * ( f(A, xk2, y) - fmin + mu/2 * normsq )            #return V

"""Use results from last simulation of the Nesterov method for strongly convex functions!"""

# Load matrix A
A = np.load('data_A.npy')

# Load corresponding data points {y_i}
y = np.load('data_b.npy')

n = A.shape[0]           #size of the data sets
order = A.shape[1]       #order of the poly fit
u = A[:,1]               #extract the u's from the A matrix
assert len(y) == len(u)  #ensure that the data sets are the same length

print("Size of data sets: ", n)
print("Order of fit: ", order)

AtA = np.matmul(A.T, A)  #compute the constant value of A^T.A once here
Aty = np.matmul(A.T, y)  #compute the constant value of A^T.y once here

xmin = nesterovStrongEst3   #load the numerical estimate for the minimiser from the previous section
fmin = f(A, xmin, y)        #calculate the numerical minimum value of f 

"""define necessary numerical constants"""

lam = 0.01 * np.linalg.norm( AtA )   #compute lambda, the regularisation parameter
print('lam: ', lam)

rtol = 1e-2      #relative tolerance - same tolerances as the previous section here!
atol = 1e-4      #absolute tolerance - since they are the same, we will stop computing V when we reach the same minimum as before
                 #                   - this way, we won't see any apparent issues from going further than this numerical minimum (which isn't the true minimum)
                 #                   - if we went further than this numerical minimum, we might see that the inequality doesn't hold

"""determine suitable L and mu for this strongly convex problem"""

hess = AtA + np.identity(order)*lam   #the Hessian matrix now equals to A^T.A + lam.I
eigs = np.linalg.eigvals( hess )      #compute the eigenvalues of the Hessian matrix
L = np.max( eigs ).real               #Lipschitz gradient constant is the maximum eigenvalue
mu = np.min( eigs ).real              #continuously differentiable constant for strong convexity is the minimum eigenvalue

print('L = ', L)
print('mu = ', mu)

"""now implement the Nesterov method for strongly convex functions"""

delta = np.sqrt(mu/L)  #compute \delta

gamma = (np.sqrt(L)-np.sqrt(mu))/(np.sqrt(L)+np.sqrt(mu))   #calculate the constant value of gamma

x = np.zeros(order)                                 #load initial estimates for x0 and z0 as an array of zeros (z_k = y_k from notes as y already denotes the input data)
z = np.zeros(order)                                 #can see from formula that gamma_{-1} = 0 so x0 = z0

gradient = grad2([AtA, Aty], z, y)       #load the initial value for the gradient - note this could also be initialised to array of 1s here
targetGrad = np.zeros_like(gradient)     #create a matrix of 0s like the gradient as
                                         #want every component of the gradient to be sufficiently close to zero
    
Vs = []    #empty array to append sequential values of V
k = 1      #need to keep track of the index k now

"""Use the Nesterov method for strongly convex functions here"""

while not (np.allclose(targetGrad, gradient, rtol, atol)):   #Stops if absolute(gradient) <= (atol + rtol * absolute(gradient))
                                                             #will keep iterating until the gradient of the residual...
                                                             #...is sufficiently close to zero

    gradient = grad2([AtA, Aty], z, y)   #compute next value of gradient
                                         #now update the sequences x and z
    xNew = z - 1./L*gradient             #cannot just overwrite x with next iteration as the sequence z...
    z = xNew + gamma*(xNew - x)          #...depends on both x_{k} and x_{k+1}

    Vs.append( V(A, xNew, x, k, xmin, fmin, y) )   #append this iteration of V(x_k, x_{k-1})
    
    x = xNew                             #now overwrite x with xNew
    k += 1                               #increment the index by one
    
"""numerically checking that the sequence of Vs is monotonically decreasing"""
    
truth = True                                                #assume that the sequence is monotonically decreasing
for idx in range(len(Vs)-1):                                #loop over all the iterations of V
    if Vs[idx+1] > Vs[idx]:                                 #If a V is larger than the previous V...
        print('Problem! Statement untrue!', idx, Vs[idx])   #...print this error message
        truth = False                                       #...and change our test boolean

if truth:                                                   #if our test boolean is still true...
    print('\nChecked numerically - Statement is true!')     #...print out this confirmation message 

"""plot V_i/V_{i-1} which should be less than 1 for each V"""    

plt.plot(np.linspace(1, len(Vs), len(Vs)-1), np.divide(Vs[1:],Vs[:-1]))   #plot V_i/V_{i-1}
plt.hlines(1, 0, 80, 'k', '--')                                           #plot a horizontal line at y = 1. Need V_i/V_{i-1} to be below this line
plt.xlabel('Iteration')
plt.ylabel('$V_{i} \quad / \quad V_{i-1}$')
plt.title('Checking that $V(x_{k+1},x_{k}) \leq V(x_{k},x_{k-1})$')
plt.show()
```
